```python
import pandas as pd
```

```python
df = pd.read_csv('/content/WA_Fn-UseC_-Telco-Customer-Churn.csv')
```

**MISSING VALUES CHECKING**

```python
df.head()
```

```python
# Pengecekan tipe data pada setiap kolom

df.dtypes
```

```python
# Mengubah data type 'TotalCharges' menjadi float 

df['TotalCharges'] = pd.to_numeric(df['TotalCharges'],errors='coerce')
df.dtypes
```

```python
# Mengecek missing values pada kolom

df.isnull().sum()
```

```python
# Melakukan pengecekan baris mana yang memiliki missing values 

nan_rows = df[df.isnull().T.any().T]
nan_rows
```

```python
# Mengganti nilai yang hilang dengan 0, karena total charges dihitung dari tenur*monthly charges
# Misalkan df_1 sebagai dataset yang sudah tidak ada missing value

df_1 = df.fillna(value=0)
df_1
```

```python
# Melakukan pengecekan pada salah satu baris yang ada missing values, untuk memastikan apakah nilai sudah terganti atau belum

df_1[488:489]
```

**CATEGORICAL DATA ENCODING**

```python
df_1
```

```python
# Menghapus kolom 'customerID'

del df_1['customerID']
df_1
```

```python
# Membuat variabel dummy untuk kolom yang memiliki tipe data string

df_dummy = pd.get_dummies(df_1)
df_dummy
```

 **ANOMALIES AND OUTLIERS**



```python
# Membuat Boxplot

import matplotlib.pyplot as plt
plt.boxplot(df_dummy['TotalCharges'])
plt.show()
```

```python
# Menghitung Outlier

Q1 = df_dummy['TotalCharges'].quantile(0.25)
Q3 = df_dummy['TotalCharges'].quantile(0.75)
IQR = Q3 - Q1 
lower_bound = Q1 - 1.5*IQR
upper_bound = Q3 + 1.5*IQR

outliers = df_dummy[df_dummy['TotalCharges']>upper_bound]
outliers
```
